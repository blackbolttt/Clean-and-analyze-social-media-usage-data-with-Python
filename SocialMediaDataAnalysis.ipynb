```python
df = pd.read_csv('SocialMediaUsersDataset.csv').set_index("UserID")
```

```python
df = df.drop(columns="Name")
```

```python
df
```

```python
df.dtypes
```

```python
df["Interests"] = df["Interests"].str.replace("'","").str.replace(",","").str.replace("and ","")
```

```python
df
```

```python
data = df['Interests'].str.get_dummies(sep=' ').max(level=0).astype(bool).astype(int)
```

```python
data
```

```python
df_updated = pd.merge(df.drop(columns="Interests"),data,on='UserID')
```

```python
df_updated.dtypes
```

```python
df_updated['Age'] = 2024 - pd.to_datetime(df['DOB']).dt.year
```

```python
df_updated = df_updated.drop(columns = ["DOB","City"])
```

```python
df_updated
```

```python
import country_converter as coco
```

```python
# Initialize the converter
cc = coco.CountryConverter()

# Map countries to continents
def get_continent(country):
    try:
        return cc.convert(names=country, to='continent')
    except:
        return None
```

```python
df_updated['continents'] = df_updated['Country'].apply(get_continent)
```

```python
df_updated['continents'].value_counts()
```

```python
df_updated['continents'].replace({'Africa': 'Africa and Others', 'Oceania': 'Africa and Others', 'Antarctica': 'Africa and Others'}, inplace=True)
```

```python
df_updated['continents'].value_counts()
```

```python
df_updated = df_updated.drop(columns="Country")
```

```python
df_updated
```

```python
info = ['Gender', 'Age', 'continents']

# Identify the rest of the columns as interests
interests = [col for col in df_updated.columns if col not in info]
```

```python
print(info,interests)
```

```python
# Calculate the count of each gender category
gender_counts = df_updated['Gender'].value_counts()

# Plotting using matplotlib
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color='skyblue', edgecolor='black')  # Bar plot
plt.title('Count of Each Gender Category')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Keep x-axis labels vertical
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Aggregate sum of interests based on gender
interests_sum = df_updated.groupby('Gender')[interests].sum().reset_index()
print(interests_sum)
# Reshape the data for plotting
interests_melted = interests_sum.melt(id_vars='Gender', var_name='Interest', value_name='Count')
print(interests_melted)
```

```python
# Plotting using seaborn
plt.figure(figsize=(16, 10))
sns.barplot(x='Interest', y='Count', hue='Gender', data=interests_melted)
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.title('Interests by Gender')
plt.ylabel('Total Count')
plt.xlabel('Interests')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
print(df_updated["Age"].min(),df_updated["Age"].max())
```

```python
# Define age ranges
bins = [19, 30, 45, 60, 100]
labels = ['19-30', '31-45', '46-60', '60+']
```

```python
# Categorize Age into these bins

df_updated['AgeRange'] = pd.cut(df_updated['Age'], bins=bins, labels=labels, right=False)
```

```python
# Aggregate sum of interests based on age ranges
interests_sum = df_updated.groupby('AgeRange')[interests].sum().reset_index()

# Reshape the data for plotting
interests_melted = interests_sum.melt(id_vars='AgeRange', var_name='Interest', value_name='Count')
```

```python
# Plotting using seaborn
plt.figure(figsize=(16, 10))
sns.barplot(x='Interest', y='Count', hue='AgeRange', data=interests_melted)
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.title('Interests by Age Range')
plt.ylabel('Total Count')
plt.xlabel('Interests')
plt.legend(title='Age Range')
plt.tight_layout()
plt.show()
```

```python
# Aggregate count of records based on AgeRange and Gender
age_gender_counts = df_updated.groupby(['AgeRange', 'Gender']).size().reset_index(name='Count')
```

```python
# Plotting using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='AgeRange', y='Count', hue='Gender', data=age_gender_counts)
plt.title('Distribution of Age Ranges Across Genders')
plt.ylabel('Count')
plt.xlabel('Age Range')
plt.tight_layout()
plt.show()
```

```python
# Group by 'continents' and aggregate the sum for each interest
interests_by_continent = df_updated.groupby('continents')[interests].sum().reset_index()

# Melt the DataFrame for plotting
melted_data = pd.melt(interests_by_continent, id_vars=['continents'], value_vars=interests, var_name='Interest', value_name='Count')
```

```python
# Plot using seaborn
plt.figure(figsize=(15, 10))
sns.barplot(x='Interest', y='Count', hue='continents', data=melted_data, ci=None)
plt.title('Interests Based on Continents')
plt.xlabel('Interests')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.legend(title='Continents')
plt.show()
```
